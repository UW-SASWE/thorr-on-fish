```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
proj_dir = Path("../../../")
```

```python
col_300rkm = pd.read_csv(proj_dir / "Code/notebooks/EDA/Columbia_River_31.csv")
dart_tddo = pd.read_csv("/Users/gdarkwah/Library/CloudStorage/OneDrive-UW/01-Research/02-Dams, Temperature and Fish/Data/insitu/conditions/processed/DART_TDDO.csv")
```
